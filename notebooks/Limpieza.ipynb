```python
# Ahora se creran los pesos maximos para cada divión de cada pelea y se creara una columna nueva llamada "PESO"
peso = []

for indice, fila in dff.iterrows():

    if fila["WEIGHT CLASS"] == "Women's Strawweight":
        peso.append(115)
    elif fila["WEIGHT CLASS"] in ["Flyweight", "Women's Flyweight"]:
        peso.append(125)
    elif fila["WEIGHT CLASS"] in ["Bantamweight", "Women's Bantamweight"]:
        peso.append(135)
    elif fila["WEIGHT CLASS"] in ["Featherweight", "Women's Featherweight" ]:
        peso.append(145)
    elif fila["WEIGHT CLASS"] == "Lightweight":
        peso.append(155)
    elif fila["WEIGHT CLASS"] == "Welterweight":
        peso.append(175)
    elif fila["WEIGHT CLASS"] == "Middleweight":
        peso.append(185)
    elif fila["WEIGHT CLASS"] == "Light Heavyweight":
        peso.append(205)
    elif fila["WEIGHT CLASS"] == "Heavyweight":
        peso.append(265)
    elif fila["WEIGHT CLASS"] == "Super Heavyweight":
        peso.append("unlimited")
    elif fila["WEIGHT CLASS"] == "Open Weight":
        peso.append("open")
    elif fila["WEIGHT CLASS"] == "Catch Weight":
        peso.append("unknow")
dff["PESO"] = peso
```
